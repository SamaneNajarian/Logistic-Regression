### ** ðŸ’¼ Dataset Description

This dataset contains customer-level information related to a marketing or promotional campaign. Each row represents one customer, and the following variables are included:

Variables

no (Integer)
A unique identifier for each record.

Purpose: Used only for indexing and does not carry analytical meaning.

Channel (Categorical / String)
The marketing or communication channel through which the customer was contacted.

Example values: Mail, Online, Phone

Purpose: Helps analyze the effectiveness of different marketing channels.

IFBundle (Categorical / String)
Indicates whether the customer received a bundled offer.

Example values: Bundle, NoBundle

Purpose: Used to study the impact of bundled offers on customer behavior.

IFUseCoupon (Categorical / String)
Shows whether the customer used a discount coupon.

Example values: Yes, No

Purpose: Useful for evaluating coupon usage behavior and the effectiveness of promotional incentives.

```r
summary(data)
levels(data$IFBundle)
data$IFBundle <- factor(data$IFBundle, levels = c("NoBundle","Bundle"))
levels(data$IFBundle)
```

```r
table(data$IFUseCoupon, data$Channel)
cat("\n---------------------------------------\n")

table(data$IFUseCoupon, data$IFBundle)
cat("\n---------------------------------------\n")

table(data$Channel, data$IFBundle)
cat("\n---------------------------------------\n")

table(data$Channel, data$IFBundle, data$IFUseCoupon)
cat("\n---------------------------------------\n")
```

```r
##Cross Tabulation Analysis----------------------
cross_tab <- table(data$IFBundle, data$IFUseCoupon)
cross_tab
```

```r
margin.table(cross_tab, 1) #over rows
margin.table(cross_tab, 2) #over columns
```

```r
prop.table(cross_tab)
prop.table(cross_tab, 1) #over rows
prop.table(cross_tab, 2) #over columns
```

```r
#Chi-Square Test
#H0: IFUseCoupon is independent of IFBundle
#H1: IFUseCoupon is related to IFBundle
#If p-value < 0.05 reject H0
chisq.test(cross_tab)
```

```r
data$IFUseCoupon <- factor(data$IFUseCoupon,
                           levels = c("No","Yes"))
```

# <p style="font-family:newtimeroman;color:#FFD700;font-size:90%;text-align:center;border-radius:10px 10px;">Modeling</p>

```r
##General Linear Model (GLM)---------------------
##model1-----------------------------------------
m1 <- glm(IFUseCoupon ~ IFBundle, data = data, family = "binomial")
summary(m1)
```

```r
#Overall fit-------------------------------------
#deviance statistic
#H0: the model is not better than chance at predicting the outcome
#p-value < 0.05 reject H0

modelChi1 <- m1$null.deviance - m1$deviance
Chidf1    <- m1$df.null - m1$df.residual
Chisq_prob1 <- 1 - pchisq(modelChi1, Chidf1)
Chisq_prob1
```

```r
m2 <- glm(IFUseCoupon ~ IFBundle + Channel, data = data, family = "binomial")
summary(m2)
```

```r
#why negative contribution of the promotion bundle?
exp(coef(m2))
exp(confint(m2))
```

```r
m3 <- glm(IFUseCoupon ~ IFBundle + Channel + IFBundle : Channel, data = data, family = "binomial")
summary(m3)
```

```r
#Overall fit-------------------------------------
#deviance statistic
#H0: the model is not better than chance at predicting the outcome
#p-value < 0.05 reject H0

modelChi3 <- m3$null.deviance - m3$deviance
Chidf3    <- m3$df.null - m3$df.residual
Chisq_prob3 <- 1 - pchisq(modelChi3, Chidf3)
Chisq_prob3
```
